```python
import numpy as np
import os
import matplotlib.pyplot as plt
from PIL import Image
from skimage.util.shape import view_as_windows
```

```python
def check_noise(global_image, local_image, mu, gamma):
    if abs(global_image.mean() - local_image.mean()) >= (mu * local_image.mean()):
        return False

    if abs(global_image.var() - local_image.var()) >= (gamma * local_image.var()):
        return False
    
    return True

            
def generate_noise_patch(image, normalize=True,
                         global_size = 64, global_stride=32,
                         local_size=16, local_stride=16,
                         mu=0.1, gamma=0.25):
    row, col, ch = image.shape
    
    noise_patch = np.empty((1, global_size, global_size, ch), image.dtype)
    image_patch = np.empty((1, global_size, global_size, ch), image.dtype)
    
    global_patches = view_as_windows(image, (global_size, global_size, ch), step=global_stride)
    global_patches = global_patches.reshape(-1, global_size, global_size, ch)
    
    for global_image in global_patches:
        noise = True
        
        local_patches = view_as_windows(global_image, (local_size, local_size, ch), step=local_stride)
        local_patches = local_patches.reshape(-1, local_size, local_size, ch)
        
        for local_image in local_patches:
            noise = check_noise(global_image, local_image, mu, gamma)
            
            if not noise: break
        
        if not noise:                
            image_patch = np.concatenate((image_patch, np.expand_dims(global_image, axis=0)), axis=0)
        else:
            noise_patch = np.concatenate((noise_patch, np.expand_dims(global_image, axis=0)), axis=0)
    
    if normalize:
        image_patch -= image_patch.mean()
        noise_patch -= noise_patch.mean()
    
    return image_patch[1:,], noise_patch[1:,]
```

```python
original = Image.open("/home/col/Desktop/BME 590/HW3/gcbd/noise_generate/60079.jpg")
display(original)
```

```python
original = np.array(original)
image, noise = generate_noise_patch(original, normalize = False)
```

```python
fig, axs = plt.subplots(2, 3)
fig.set_size_inches(15., 10.)

for i in range(2):
    for j in range(3):
        axs[i,j].imshow(noise[i*3+j,])
        axs[i,j].axis('off')
```
